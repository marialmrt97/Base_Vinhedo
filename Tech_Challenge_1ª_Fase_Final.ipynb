## Importa dado de Exportação

```python
import pandas as pd
dados = pd.read_csv('/content/ExpVinho (1).csv', encoding='utf-8',  sep=';', thousands='.', decimal=',')
dados.head()
```

```python
novos_nomes_colunas = {}
for coluna in dados.columns:
    if coluna not in ["Id", "País"]:
        if ".1" in coluna:
            novo_nome = coluna.replace(".1", "") + "-$"
        else:
            novo_nome = coluna + "-kg"
        novos_nomes_colunas[coluna] = novo_nome
    else:
        novos_nomes_colunas[coluna] = coluna
dados.rename(columns=novos_nomes_colunas, inplace=True)
dados.head()
```

***## Trás Somente o que contém $ pra analisar***

```python
import pandas as pd

dados_vendas = dados.copy()
colunas_vendas = ['País'] + [coluna for coluna in dados_vendas.columns if coluna.endswith("-$")]
dados_vendas = dados_vendas[colunas_vendas]
dados_vendas.columns = [coluna.replace("-$", "") for coluna in dados_vendas.columns]
dados_vendas = dados_vendas.melt(id_vars='País', var_name='Ano', value_name='USD Exportados')
dados_vendas['País_de_Origem'] = 'Brasil'
dados_vendas['Posição'] = range(1, len(dados_vendas) + 1)
dados_vendas = dados_vendas.sort_values(by='USD Exportados', ascending=False)
dados_vendas = dados_vendas.set_index('Posição')
dados_vendas.head()
```

```python
path = '/content/drive/My Drive/FIAP_5DTAT_Grupo_59/dados_vendas.csv'  # Ajuste o caminho conforme necessário

# Salvar o DataFrame como CSV no caminho especificado
dados_vendas.to_csv(path, sep=";", index=False)  # 'index=False' para não escrever os índices no arquivo CSV
```

***Montante em Litros exportado nos últimos 15 anos por país***

```python
dados_qtd = dados.copy()
colunas_qnt = ['País'] + [coluna for coluna in dados_qtd.columns if coluna.endswith("-kg")]
colunas_qtd = dados[colunas_qnt]
colunas_qtd.columns = [coluna.replace("-kg", "") for coluna in colunas_qtd.columns]
colunas_qtd = colunas_qtd.melt(id_vars='País', var_name='Ano', value_name='Litros')
colunas_qtd['Posição'] = range(1, len(colunas_qtd) + 1)
colunas_qtd = colunas_qtd.sort_values(by='Ano', ascending=False)
colunas_qtd = colunas_qtd.set_index('Posição')
colunas_qtd.head()

path = '/content/drive/My Drive/FIAP_5DTAT_Grupo_59/colunas_qtd.csv'  # Ajuste o caminho conforme necessário

# Salvar o DataFrame como CSV no caminho especificado
colunas_qtd.to_csv(path, sep=";",index=False)  # 'index=False' para não escrever os índices no arquivo CSV
```

***TABELA DE MONTANTE DE VENDAS (US$) DE EXPORTAÇÃO NOS ÚLTIMOS 15 ANOS POR PAÍS***

```python
dados_qtd = dados.copy()

colunas_valor = ['País'] + [coluna for coluna in dados_qtd.columns if coluna.endswith("-$")]
dados_ultimos_15_anos = dados_qtd[colunas_valor].copy()

#Calcula vendas por país
dados_ultimos_15_anos['Total_US$'] = dados_ultimos_15_anos.iloc[:, 1:].sum(axis=1)

#Filtra os países que tem vendas >0
dados_filtrados = dados_ultimos_15_anos[dados_ultimos_15_anos['Total_US$'] > 0].copy()

#Ordena
dados_ordenados = dados_filtrados.sort_values(by='Total_US$', ascending=False)

#Cria coluna de ranking
dados_ordenados['Posição'] = range(1, len(dados_ordenados) + 1)
#Renomeia as colunas e cria coluna de país de origem
dados_ordenados = dados_ordenados.rename(columns={'País': 'País_de_Destino'})
dados_ordenados['País_de_Origem'] = 'Brasil'
dados_ordenados['Total_US$'] = dados_ordenados['Total_US$'].apply(lambda x: "${:,.2f}".format(x))
dados_ordenados[['País_de_Origem', 'País_de_Destino', 'Total_US$', 'Posição']].head(100)
```

***TABELA DE MONTANTE DE LITROS EXPORTADOS (1KG = 1L) NOS ÚLTIMOS 15 ANOS***

```python
# Calcula o montante de exportação por país em kg (1L) nos últimos 15 anos
ultimos_15_anos_qtd = ['País'] + [str(ano) + "-kg" for ano in range(2008, 2023)]
dados_ultimos_15_anos_qtd = dados[colunas_qnt].copy()
dados_ultimos_15_anos_qtd.loc[:, 'Total_kg'] = dados_ultimos_15_anos_qtd.iloc[:, 1:].sum(axis=1)
# Filtra os dados para incluir apenas os países com exportações maiores que zero
dados_filtrados_qtd = dados_ultimos_15_anos_qtd[dados_ultimos_15_anos_qtd['Total_kg'] > 0].copy()
# Organiza os dados por maior valor em kg
dados_ordenados_qtd = dados_filtrados_qtd.sort_values(by='Total_kg', ascending=False)
# Cria a coluna de ranking de posição
dados_ordenados_qtd['Posição'] = range(1, len(dados_ordenados_qtd) + 1)
# Renomeia as colunas
dados_ordenados_qtd = dados_ordenados_qtd.rename(columns={'País': 'País_de_Destino'})
dados_ordenados_qtd = dados_ordenados_qtd.rename(columns={'Total_kg': 'Total_Litros'})
dados_ordenados_qtd['País_de_Origem'] = 'Brasil'

dados_ordenados_qtd['Total_Litros'] = dados_ordenados_qtd['Total_Litros'].apply(lambda x: "{:,.2f}".format(x))
dados_ordenados_qtd[['País_de_Origem', 'País_de_Destino', 'Total_Litros', 'Posição']].head(100)
```

***TABELA CONTENDO PAÍS DE ORIGEM, DESTINO, US$ E LITROS EXPORTADOS (1KG = 1L)***

```python
## Juntas tabelas de vendas com tabela de quantidade
dados_combinados = pd.merge(dados_ordenados, dados_ordenados_qtd, on=['País_de_Origem', 'País_de_Destino'], how='outer')
dados_final = dados_combinados[['País_de_Origem', 'País_de_Destino', 'Total_US$', 'Total_Litros']]
dados_final = dados_final.rename(columns={'Total_US$': 'Total Exportado (US$)', 'Total_Litros': 'Total Exportado (L)'})
dados_final['Posição'] = range(1, len(dados_final) + 1)
dados_final = dados_final.set_index('Posição')
dados_final.head(100)
```
